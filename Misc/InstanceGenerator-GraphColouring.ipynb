```python
import os
import csv
import time
import random
import networkx as nx
```

```python
## Random generation of Color graphs + Solve and Save
count = 0
randomIGClabels = []


os.makedirs(os.path.join("LFvsDSATUR","LF"), exist_ok=True)
os.makedirs(os.path.join("LFvsDSATUR","DSATUR"), exist_ok=True)

with open("RandomgGraphColLabels-LFvsDSATUR.csv", "w") as resultFile:
  resultFile.write("FileName,Label,Colours0,Time0,Colours1,Time1")


for r in range(0,1000):
  for i in range(10,200):
    #Generate
    edges = random.randint(i,(i*i)//2)
    randomGraph= nx.generators.random_graphs.gnm_random_graph(i,edges)
    randomGraph.remove_edges_from(nx.selfloop_edges(randomGraph))
    randomGraph.name = str(i) + "-" + str(edges) + "r" + str(r)

    label = 0
    folder = ""

    #Solve
    start = time.time()
    colors0 = nx.coloring.greedy_color(randomGraph, strategy="largest_first")
    time0 = time.time() - start

    start = time.time()
    colors1 = nx.coloring.greedy_color(randomGraph, strategy="saturation_largest_first")
    time1 = time.time() - start

    #Label
    if(max(colors0.values()) == max(colors1.values())):
      if(time1>time0):      
        label = 0
        folder = "LF"
      else:
        label = 1
        folder =  "DSATUR"
        count += 1 
    elif(max(colors0.values()) < max(colors1.values())):
      label = 0
      folder = "LF"
    else:
      label = 1
      folder =  "DSATUR"
      count += 1 

    # Save
    with open("LFvsDSATUR/" + folder + "/" + randomGraph.name + ".csv" , 'w', newline='\n') as instanceFile:
      for line in nx.generate_edgelist(randomGraph, data=False,delimiter=','):
        instanceFile.write(line + "\n")
    with open("RandomgGraphColLabels-LFvsDSATUR.csv", "a") as resultFile:
      rw = csv.writer(resultFile, quoting=csv.QUOTE_ALL)
      rw.writerow((randomGraph.name,label,max(colors0.values()), time0, max(colors1.values()),time1 ))

print(count)
```

```python
train_data = load_files("balanced-LFvsDSATUR", shuffle=True)
c = list(zip(train_data.data, train_data.target))

random.shuffle(c)
a, b = zip(*c)

x_train = np.array(a)[:50000]
y_train = np.array(b)[:50000] 
x_test = np.array(a)[50000:70000]
y_test = np.array(b)[50000:70000]
```

```python
a,c,b,train_data = [],[],[],[]
```

```python
# Initialize the text regressor.
reg = ak.TextRegressor(overwrite=False, max_trials=1)  
# Feed the text regressor with training data.
reg.fit(x_train, y_train, epochs=5)
# Predict with the best model.
predicted_y = reg.predict(x_test)
```

```python
print(reg.evaluate(x_test, y_test))
```

```python
input_node = ak.TextInput()
output_node = ak.TextBlock(block_type="sequence")(input_node)
output_node = ak.ClassificationHead()(output_node)
clf = ak.AutoModel(
    inputs=input_node, outputs=output_node, overwrite=False, max_trials=1
)
clf.fit(x_train, y_train, epochs=100)
```

```python
predicted_y = clf.predict(x_test)
print(clf.evaluate(x_test, y_test))
print(classification_report(y_test, predicted_y))
```
