```python
import pandas as pd
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import shutil
```

```python
df = pd.read_csv('selectedset-TSP.csv')
plt.figure(dpi=800)
plt.tight_layout()
fig, axes = plt.subplots(ncols=2,figsize=(36, 9))

df = df.dropna()
df["Tdiff"] = df["Time0"]-df["Time1 "]
df["TourDiff"] = df["TourLength0"] - df["TourLength1"]

lab0 = df['newLabel'].tolist()
winner0 = []
winner1 = []
size = 20
colors0 = []
for i,l in enumerate(lab0):
  if l == 0:
    winner0.append(size)
    winner1.append(0) 
    colors0.append('green') 
  else:
    winner0.append(0)
    winner1.append(size)
    if df['TourDiff'].iloc[i] > 0 and df['Time0'].iloc[i] < 0.1:
      colors0.append('blue') 
    else:
      colors0.append('red') 



#df["Tdiff"].hist(bins=20,rwidth=0.9,color="g",ax=axes[0])
df.plot.scatter("Time0","Time1 ", s=winner0, c=colors0, marker ="x", alpha=0.3, edgecolors='none', ax=axes[0])
df.plot.scatter("Time0","Time1 ", s=winner1, c=colors0, marker ="s", alpha=0.3, edgecolors='none', ax=axes[0])
df.plot.scatter("TourLength0","TourLength1", s=winner0, c=colors0, marker ="x", alpha=0.4, edgecolors='none', ax=axes[1])
df.plot.scatter("TourLength0","TourLength1", s=winner1, c=colors0, marker ="s", alpha=0.4, edgecolors='none', ax=axes[1])
axes[0].set_xlabel('CHR Time',fontsize=20)
axes[0].set_ylabel('GREEDY Time',fontsize=20)
axes[0].tick_params(axis="x", labelsize=16)
axes[0].tick_params(axis="y", labelsize=16)
axes[1].set_xlabel('CHR Tour Length',fontsize=20)
axes[1].set_ylabel('GREEDY Tour Length',fontsize=20)
axes[1].tick_params(axis="x", labelsize=16)
axes[1].tick_params(axis="y", labelsize=16)

fig.suptitle("TSP Instances", fontsize=24)

#axes[0].set_yscale('log')
#axes[0].set_xscale('log')
#axes[1].set_yscale('log')
#axes[1].set_xscale('log')
#sns.lmplot(data=df,x="Time0",y="Time1 ",line_kws={'color': 'g'})
#df["TourLength0"].corr(df["TourLength1"])**2
#df.describe()

 
#plt.show()
plt.savefig("TSPInstances.png",bbox_inches='tight', pad_inches=0.1)
#df['newLabel'].value_counts()
```

```python
df = pd.read_csv('GC-Results-LFvsDSATUR.csv')
plt.figure(dpi=800)
plt.tight_layout()
fig, axes = plt.subplots(ncols=2,figsize=(36, 9))

df = df.dropna()
df["Tdiff"] = df["Time0"]-df["Time1"]
df["CDiff"] = df["Colours0"] - df["Colours1"]

lab0 = df['Label'].tolist()
winner0 = []
winner1 = []
size = 20
colors0 = []
for i,l in enumerate(lab0):
  if l == 0:
    winner0.append(size)
    winner1.append(0) 
    if df['CDiff'].iloc[i] == 0:
      colors0.append('red')
    else:
      colors0.append('green')

  else:
    winner0.append(0)
    winner1.append(size)
    if df['CDiff'].iloc[i] == 0:
      colors0.append('cyan') 
      print("happened")
    else:
      colors0.append('blue') 

df.plot.scatter("Time0","Time1", s=winner1, c=colors0, marker ="o", alpha=0.4, edgecolors='none', ax=axes[0])
df.plot.scatter("Time0","Time1", s=winner0, c=colors0, marker ="^", alpha=0.4, edgecolors='none', ax=axes[0])
df.plot.scatter("Colours0","Colours1", s=winner0, c=colors0, marker ="^", alpha=0.4, edgecolors='none', ax=axes[1])
df.plot.scatter("Colours0","Colours1", s=winner1, c=colors0, marker ="o", alpha=0.4, edgecolors='none', ax=axes[1])
axes[0].set_xlabel('LF Time',fontsize=20)
axes[0].set_ylabel('DSATUR Time',fontsize=20)
axes[0].tick_params(axis="x", labelsize=16)
axes[0].tick_params(axis="y", labelsize=16)
axes[1].set_xlabel('LF Colours',fontsize=20)
axes[1].set_ylabel('DSATUR Colours',fontsize=20)
axes[1].tick_params(axis="x", labelsize=16)
axes[1].tick_params(axis="y", labelsize=16)
fig.suptitle("Graph Colouring Instances", fontsize=24)

axes[0].set_yscale('log')
axes[0].set_xscale('log')
axes[1].set_yscale('log')
axes[1].set_xscale('log')

plt.savefig("GCInstances.png",bbox_inches='tight', pad_inches=0.1)
```
