```python
def extractTSPparams(file_path):
  instance = []
  lines = []
  #problem data start at line 6
  start = 6
  with open(file_path) as f:
    lines = f.read().splitlines()

  for i in range(start, len(lines)-1):
    line = lines[i].split( )
    instance.append([int(float(line[1])),int(float(line[2]))])
  #print(instance)
  return instance
```
